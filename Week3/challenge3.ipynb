```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
mushrooms = pd.read_csv('mushrooms.csv')

cols = mushrooms.select_dtypes(["object"]).columns
mushrooms[cols] = mushrooms[cols].astype('category')
edibleclass=mushrooms.groupby(['cap-color']).count()
edibleclass
labels=['Brown','Buff','Cinnamon','Green','Pink','Purple','Red','White','Yellow']
plt.pie(edibleclass['population'],labels=labels,autopct='%.1f %%')
plt.title('cap-color')
plt.show()
gill=mushrooms.groupby(['gill-attachment']).count()
gill
labels=['Attached','Free']

plt.pie(gill['class'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85)
  
center_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('gill-attachment')
  
plt.show()
```
