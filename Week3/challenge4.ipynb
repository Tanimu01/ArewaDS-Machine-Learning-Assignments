```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('honey.csv')
honey['xn']=honey['state']
honey['xn']
sns.relplot(x="prodvalue", y=honey['xn'], data=honey, height=15, aspect=.5);
sns.relplot(x="stocks",y=honey['year'], data=honey, height=15, aspect=.5)
sns.relplot(x="stocks", y="year", hue="year", palette="YlOrBr", data=honey, height=15, aspect=.5);
sns.relplot(x="year", y="prodvalue", kind="line", data=honey);
sns.relplot(x="year", y="stocks", kind="line", data=honey);
```
